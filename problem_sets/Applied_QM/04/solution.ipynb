## a)

For a particle under the magnetic field, we can write the perturbation as:

$$\delta H = eE z \sin(\omega t) \quad \Rightarrow \quad H = H_0 + \delta H$$

where $H_0$ is the unperturbed Hamiltonian. Where $|i\rangle$ is the initial state and $|f\rangle$ is the final state, by replacing from the problem statement, we get:

$$\langle f|\delta H|i\rangle = eE_0\langle f|z\sin(\omega t)|i\rangle = eE_0\langle p|z\sin(\omega t)|s\rangle$$

where $E_0$ is the electric field amplitude. Now, we can write the explicit general form for the $s$ and $p$ states:

$$\langle f|\delta H|i\rangle = eE_0\langle n`m`l`|z\sin(\omega t)|nml\rangle$$

where $nml$ are the quantum numbers for the $s$ state and $n`m`l`$ are the quantum numbers for the $p$ state. Since we need the matrix element for the $z$ operator, we can write:

$$\langle n'm'l'|[\hat L_z, \hat z] |nml\rangle = (m-m')\langle n'm'l'|z|nml\rangle$$

where we have used the fact that $[\hat L_z, \hat z] = i\hbar \hat L_z$. Thus, we notice that all the transitions get to zero unless $m = m'$. And the $s$ state has $m = 0$, therefore we can infer that the only possible transition is from $m` = 0$ to $m = 0$. Aditionally, we are only accepting transitions from $n` = n+1$, therefore we get only the state:

$$\ket{f} = \ket{210}$$

## a)
Given the Hamiltonian $H=H_0 + H'$, we want to find the matrix elements of $\braket{i|H|j}$. First, we need to express the Hamiltonian in the basis $\ket{0}$ and $\ket{1}$. The Hamiltonian can be written as a 2x2 matrix where the basis vectors are $\ket{n}$ and $\ket{\bar{n}}$. This gives:

$$H = \begin{bmatrix} \braket{n|H|n} & \braket{n|H|\bar{n}} \\ \braket{\bar{n}|H|n} & \braket{\bar{n}|H|\bar{n}} \end{bmatrix}$$

Let's compute each of these matrix elements individually.

1. $\braket{n|H|n} = \braket{n|H_0|n} + \braket{n|H'|n} = m_n c^2 + \epsilon$
2. $\braket{n|H|\bar{n}} = \braket{n|H_0|\bar{n}} + \braket{n|H'|\bar{n}} = \epsilon$
3. $\braket{\bar{n}|H|n} = \braket{\bar{n}|H_0|n} + \braket{\bar{n}|H'|n} = \epsilon$
4. $\braket{\bar{n}|H|\bar{n}} = \braket{\bar{n}|H_0|\bar{n}} + \braket{\bar{n}|H'|\bar{n}} = m_n c^2 + \epsilon$

So, the Hamiltonian becomes:

$$H = \begin{bmatrix} m_n c^2 + \epsilon & \epsilon \\ \epsilon & m_n c^2 + \epsilon \end{bmatrix}$$

where $I$ is the identity matrix and $\lambda$ represents the eigenvalues. This yields the equation:

$$\text{det}(H - \lambda I) = 0 \Rightarrow \begin{vmatrix} m_n c^2 + \epsilon - \lambda & \epsilon \\ \epsilon & m_n c^2 + \epsilon - \lambda \end{vmatrix} = 0$$

which simplifies to:

$$(m_n c^2 + \epsilon - \lambda)^2 - \epsilon^2 = 0$$

The solutions to this equation are the eigenvalues of the Hamiltonian:

$$\lambda_{\pm} = m_n c^2 + \epsilon \pm \epsilon$$

The corresponding eigenvectors can be found by substituting these eigenvalues back into the equation $(H - \lambda I)\ket{v} = 0$, and solving for the eigenvector $\ket{v}$.

For $\lambda_{+}$, the equation becomes:

$$\begin{bmatrix} -\epsilon & \epsilon \\ \epsilon & -\epsilon \end{bmatrix} \begin{bmatrix} a \\ b \end{bmatrix} = 0$$

which yields the eigenvector $\ket{v_+} = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ 1 \end{bmatrix}$.

For $\lambda_{-}$, the equation becomes:

$$\begin{bmatrix} 0 & \epsilon \\ \epsilon & 0 \end{bmatrix} \begin{bmatrix} a \\ b \end{bmatrix} = 0$$

which yields the eigenvector $\ket{v_-} = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ -1 \end{bmatrix}$.

So, the eigenvalues of the Hamiltonian are $\lambda_{+} = m_n c^2 + 2\epsilon$ and $\lambda_{-} = m_n c^2$, with corresponding eigenvectors $\ket{v_+} = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ 1 \end{bmatrix}$ and $\ket{v_-} = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ -1 \end{bmatrix}$, respectively.

This shows how the Hamiltonian matrix, representing the energy of the system, can be diagonalized to reveal the energy eigenvalues and the corresponding eigenstates.

## b)
Given that the initial state of the system at $t=0$ is a neutron $\ket{\psi(0)} = \ket{n} = \ket{0}$, we want to express this state in terms of the energy eigenstates we found earlier:

$$\ket{\psi(0)} = c_+ \ket{v_+} + c_- \ket{v_-}$$

Using the eigenvectors we found earlier, we can determine the coefficients $c_+$ and $c_-$:

$$\ket{0} = c_+ \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ 1 \end{bmatrix} + c_- \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ -1 \end{bmatrix}$$

Comparing the components, we can see that $c_+ = c_- = \frac{1}{\sqrt{2}}$. So, the initial state can be expressed as:

$$\ket{\psi(0)} = \frac{1}{\sqrt{2}}(\ket{v_+} + \ket{v_-})$$

Now, let's consider the time-evolution of the state. The time-evolved state $\ket{\psi(t)}$ can be obtained by applying the time-evolution operator, which is given by:

$$\ket{\psi(t)} = e^{-\frac{i}{\hbar}Ht}\ket{\psi(0)}$$

Since we know the energy eigenvalues $\lambda_{+}$ and $\lambda_{-}$, we can express the time-evolution operator in the energy eigenbasis:

$$\ket{\psi(t)} = \sum_{j=\pm} c_j e^{-\frac{i}{\hbar}\lambda_j t} \ket{v_j} = \frac{1}{\sqrt{2}}\left( e^{-\frac{i}{\hbar}(m_n c^2 + 2\epsilon)t} \ket{v_+} + e^{-\frac{i}{\hbar}m_n c^2 t} \ket{v_-} \right)$$

To find the probability of observing an anti-neutron at time $t$, we need to calculate the overlap between the time-evolved state $\ket{\psi(t)}$ and the anti-neutron state $\ket{\bar{n}} = \ket{1}$:

$$P_{\bar{n}}(t) = \left|\braket{1|\psi(t)}\right|^2$$

Computing the inner product:

$$\braket{1|\psi(t)} = \frac{1}{\sqrt{2}}\left( e^{-\frac{i}{\hbar}(m_n c^2 + 2\epsilon)t} \braket{1|v_+} + e^{-\frac{i}{\hbar}m_n c^2 t} \braket{1|v_-} \right)$$

Using the eigenvectors we found earlier, we have $\braket{1|v_+} = \frac{1}{\sqrt{2}}$ and $\braket{1|v_-} = -\frac{1}{\sqrt{2}}$. Substituting these values, we have:


$$\braket{1|\psi(t)} = \frac{1}{2}\left( e^{-\frac{i}{\hbar}(m_n c^2 + 2\epsilon)t} - e^{-\frac{i}{\hbar}m_n c^2 t} \right)$$

Now, to find the probability $P_{\bar{n}}(t)$, we take the absolute square of this overlap:

$$P_{\bar{n}}(t) = \left|\braket{1|\psi(t)}\right|^2 = \frac{1}{4}\left| e^{-\frac{i}{\hbar}(m_n c^2 + 2\epsilon)t} - e^{-\frac{i}{\hbar}m_n c^2 t} \right|^2$$

Expanding the absolute square, we get:

$$P_{\bar{n}}(t) = \frac{1}{4}\left( 2 - 2\cos\left(\frac{2\epsilon t}{\hbar}\right) \right)$$


## c)
The half-life is typically defined as the time it takes for the probability of a transition to reach 0.5. In our case, this corresponds to the time $t_{1/2}$ at which the probability $P_{\bar{n}}(t_{1/2}) = 0.5$.

From the equation we derived earlier, we have:

$$P_{\bar{n}}(t_{1/2}) = 0.5 = \frac{1}{4}\left(2 - 2\cos\left(\frac{2\epsilon t_{1/2}}{\hbar}\right)\right)$$

This simplifies to:

$$\cos\left(\frac{2\epsilon t_{1/2}}{\hbar}\right) = 0$$

The cosine function equals zero at $\frac{\pi}{2} + n\pi$, where $n$ is an integer. In this case, to find the first time (i.e., the smallest positive time) at which the probability equals 0.5, we should take $n=0$. Therefore, we get:

$$\frac{2\epsilon t_{1/2}}{\hbar} = \frac{\pi}{2}$$

Solving for $t_{1/2}$, we find:

$$t_{1/2} = \frac{\pi \hbar}{4 \epsilon}$$

This is the half-life of the neutron to anti-neutron transition in terms of the parameter $\epsilon$ and the reduced Planck's constant $\hbar$.

Given the infinite square well potential from $0$ to $L$, the Hamiltonian of the system is:

$$H = -\frac{\hbar^2}{2m}\nabla^2$$

The ground state wave function and ground state energy for the infinite square well are known to be:

$$\psi_0(x) = \sqrt{\frac{2}{L}} \sin\left(\frac{\pi x}{L}\right) \quad \quad E_0 = \frac{\pi^2 \hbar^2}{2mL^2}$$

Now, let's consider the trial wave function:

$$\psi(x) = x(L-x) + Cx^2(L-x)$$


Applying the Hamiltonian $H$ to the wave function $\psi(x)$:

$$
H\psi(x) = -\frac{\hbar^2}{2m} \frac{d^2}{dx^2} (x(L-x) + Cx^2(L-x))
$$

Taking the second derivative:

$$
\frac{d^2}{dx^2}(x(L-x) + Cx^2(L-x)) = -2 + 2C(1-3x)
$$

Now, we can compute the numerator of $\langle H \rangle$:

$$
\int_0^L \psi^*(x) H\psi(x) dx = -\frac{\hbar^2}{2m} \int_0^L (x(L-x) + Cx^2(L-x))(-2 + 2C(L-3x)) dx
$$

And the denominator:

$$
\int_0^L \psi^*(x) \psi(x) dx = \int_0^L (x(L-x) + Cx^2(L-x))^2 dx = \frac{C^2 L^7}{105}+\frac{C L^6}{30}+\frac{L^5}{30}
$$

Therefore we take the solution $C=0$ since we know that the state is the minimum energy state. Now, we can compute the expectation value of the Hamiltonian:

$$
\langle H \rangle = \frac{-\frac{\hbar^2}{2m} \int_0^L (x(L-x) + Cx^2(L-x))(-2 + 2C(L-3x)) dx}{\int_0^L (x(L-x) + Cx^2(L-x))^2 dx} = \frac{\hbar^2}{2m} \left(\frac{8 C^2}{C L (2 C L+7)+7}+\frac{10}{L^2}
\right)
$$

To minimize the expectation value of the Hamiltonian, we can differentiate $\langle H \rangle$ with respect to the variational parameter $C$ and set it to zero:

$$
\frac{\partial \langle H \rangle}{\partial C} = 0
$$

$$
\frac{\partial \langle H \rangle}{\partial C} = \frac{56 C (C L+2)}{(C L (2 C L+7)+7)^2}
$$

Solving for $C$ we find two solutions:

$$C=0 \quad \quad C = \frac{-2}{L}$$

The next step is to solve for $C$ to obtain the optimal value of the variational parameter. Then, we substitute this value back into the expression for $\langle H \rangle$ to obtain the energy of the ground state for the trial wave function.

$$
\langle H \rangle = \frac{\hbar^2}{2m} \Big(-\frac{8 C^2}{C L (2 C L+7)+7}-\frac{10}{L^2}\Big)\Big|_{C=0} = \frac{10 \hbar^2}{2 m L^2}
$$

Finally, to compute the error in this approximation, compare the result with the exact ground state energy for an infinite square well potential, which is:

$$
E_{1} = \frac{\hbar^2 \pi^2}{2mL^2}
$$

The error can be calculated as the absolute difference between the exact energy $E_{1}$ and the variational energy, divided by the exact energy:

$$
\text{Error} = \frac{|E_{1} - \langle H \rangle|}{E_{1}} = \Big\|\frac{10 \hbar^2}{2 m L^2} - \frac{\hbar^2 \pi^2}{2mL^2} \Big\| = \frac{\hbar^2}{2mL^2} \Big\|10 - \pi^2 \Big\|
$$

Taking for example $L=1$ and $m=1$, we get:

$$
\text{Error} = \frac{\hbar^2}{2} \Big\|10 - \pi^2 \Big\| \approx 0.0001 \hbar^2 \approx 0.1403 eV
$$

```python
import sympy as sp
import scipy.integrate as integrate
import scipy.optimize as optimize
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Constants
hbar, m, L = 1, 1, 1

def solve_variational():
    # Symbolic variables
    x, C = sp.symbols('x C')

    # Trial wave function
    psi = x*(L-x) + C*x**2*(L-x)

    # Hamiltonian applied to the wave function
    H_psi = -hbar**2/(2*m) * sp.diff(psi, x, 2)

    # Numerator and denominator of the expectation value of the Hamiltonian
    numerator = sp.integrate(psi * H_psi, (x, 0, L))
    denominator = sp.integrate(psi**2, (x, 0, L))

    # Expectation value of the Hamiltonian
    H_exp = numerator / denominator

    # Convert to a function for numerical optimization
    H_exp_func = sp.lambdify(C, H_exp)

    # Minimize the expectation value of the Hamiltonian
    res = optimize.minimize_scalar(H_exp_func)

    # Optimal variational parameter
    C_opt = res.x

    # Energy of the ground state with the trial wave function
    E_var = H_exp_func(C_opt)

    # Exact ground state energy for an infinite square well potential
    E1 = np.pi**2 * hbar**2 / (2 * m * L**2)

    error = abs(E1 - E_var) / E1

    print(f"Optimal C: {C_opt}")
    print(f"Variational Energy: {E_var}")
    print(f"Exact Energy: {E1}")
    print(f"Error: {error}")
    
solve_variational()
```

## a)
The interaction Hamiltonian in the presence of an electromagnetic field can be written as:

$$ H = -\mathbf{d} \cdot \mathbf{E} - \mathbf{\mu} \cdot \mathbf{B} $$

Here, $\mathbf{d}$ is the electric dipole moment of the atom, $\mathbf{\mu}$ is the magnetic dipole moment, $\mathbf{E}$ is the electric field, and $\mathbf{B}$ is the magnetic field. The first term $-\mathbf{d} \cdot \mathbf{E}$ represents the interaction of the atom's electric dipole moment with the electric field, and the second term $-\mathbf{\mu} \cdot \mathbf{B}$ represents the interaction of the atom's magnetic dipole moment with the magnetic field.

The electric dipole moment $\mathbf{d}$ is proportional to the charge of the electron $e$ and the typical size of the atom $a$, i.e., $\mathbf{d} \sim e \cdot a$. The magnetic dipole moment $\mathbf{\mu}$ is proportional to the Bohr magneton (which is given by $e\hbar/2m$, where $m$ is the electron mass and $\hbar$ is the reduced Planck's constant) and the typical size of the atom divided by the Bohr radius $a_0$, i.e., $\mathbf{\mu} \sim (e\hbar/2m) \cdot (a/a_0)$.

Therefore, the ratio of the magnetic interaction to the electric interaction is approximately given by:

$$ \frac{-\mathbf{\mu} \cdot \mathbf{B}}{-\mathbf{d} \cdot \mathbf{E}} \sim \frac{(e\hbar/2m) \cdot (a/a_0)}{e \cdot a} \sim \frac{\hbar}{2m a a_0} \sim \frac{v}{c} $$

where $v$ is the typical velocity of the electron in the atom, and $c$ is the speed of light. Since $v << c$, the magnetic interaction is typically much smaller than the electric interaction, and can often be neglected. This is the basis of the electric dipole approximation.


## b)

Let's write down the interaction Hamiltonian in the electric dipole approximation:

$$V = -e \mathbf{r} \cdot \mathbf{E},$$

where $ \mathbf{r} $ is the position operator, $ \mathbf{E} $ is the electric field, and $ e $ is the electron charge. We are given that $ \mathbf{E} = E_0 \boldsymbol{\epsilon} \cos(\omega t) $, where $ \boldsymbol{\epsilon} $ is a unit polarization vector, $ E_0 $ is the amplitude of the electric field, and $ \omega $ is the frequency of the radiation. Let's say that the direction of the polarization of the electric field is along the z-axis. In this case, we can write the interaction Hamiltonian as:

$$V = eE_0z \cos(\omega t),$$

where $ z $ is the z-component of the position operator. Now, let's use first-order time-dependent perturbation theory to calculate the transition rate from the ground state of the hydrogen atom to a continuum state. According to Fermi's Golden Rule, the transition rate is given by:

$$P_{fi} = \frac{{2\pi}}{{\hbar}} \left| \langle \psi_f | V | \psi_i \rangle \right|^2 \rho(E),$$

where $ | \psi_i \rangle $ is the initial state, $ | \psi_f \rangle $ is the final state, $ V $ is the interaction Hamiltonian, and $ \rho(E) $ is the density of final states. The initial state of the electron is the ground state of the hydrogen atom, which is denoted by $ | \psi_i \rangle = |1s \rangle $. The final state is a continuum state, which is denoted by $ | \psi_f \rangle = |p, \boldsymbol{\epsilon} \rangle $. Here, $ p $ is the momentum of the electron, and $ \boldsymbol{\epsilon} $ is the energy of the electron. We can write the matrix element as:

$$\langle \psi_f | V | \psi_i \rangle = eE_0 \langle \psi_f | z | \psi_i \rangle.$$

The z-component of the position operator in spherical coordinates is $ r \cos \theta $. Therefore, we need to calculate the integral:

$$\langle \psi_f | z | \psi_i \rangle = \int \psi_f^* (r \cos \theta) \psi_i dV,$$

where $ dV = r^2 \sin \theta dr d\theta d\phi $ is the volume element in spherical coordinates. The wavefunctions $ \psi_i $ and $ \psi_f $ can be written in terms of spherical harmonics and radial wavefunctions. For the hydrogen atom:

$$\psi_i = R_{10}(r)Y_{00}(\theta, \phi) = \frac{1}{\sqrt{\pi}} \left(\frac{1}{a_0}\right)^{3/2} e^{-r/a_0},$$

where $ R_{10}(r) $ is the radial wavefunction for the ground state, $ Y_{00}(\theta, \phi) $ is the spherical harmonic for $ l = 0 $ and $ m = 0 $, and $ a_0 $ is the Bohr radius. For a free electron in a continuum state:

$\psi_f = R_f(r)Y_{1m}(\theta, \phi) e^{ipr/\hbar}$,

where $R_f(r)$ is the radial wavefunction for a free electron, $Y_{1m}(\theta, \phi)$ is the spherical harmonic for $l = 1$ and $m$ can be $-1$, $0$, or $1$, and $p$ is the momentum of the electron.

In the electric dipole approximation, the electron can absorb a photon and transition from an $s$ orbital to a $p$ orbital. Therefore, we need to calculate three integrals corresponding to $m = -1$, $0$, and $1$. The matrix element $\langle\psi_f|z|\psi_i\rangle$ will be nonzero only for $m = 0$, because only the spherical harmonic $Y_{10}(\theta, \phi)$ contains a factor of $\cos \theta$, which is necessary to integrate with the operator $r \cos \theta$ in the interaction Hamiltonian. So, we can write:

$$
\langle \psi_f | z | \psi_i \rangle = eE_0 \int \psi_f^* (r \cos \theta) \psi_i dV.
$$

Substituting the expressions for $\psi_f$ and $\psi_i$, and evaluating the integral, we obtain:

$$
\langle \psi_f | z | \psi_i \rangle = eE_0 \left(\frac{1}{\sqrt{\pi}}\right)\left(\frac{1}{a_0}\right)^{\frac{3}{2}} \int R_f(r)Y_{10}(\theta, \phi) (r \cos \theta) e^{-r/a_0} dV.
$$

Next, we will need to calculate the density of final states $\rho(E)$. The density of states in momentum space for a free particle in three dimensions is given by $\rho(p) = \frac{V}{(2\pi\hbar)^3}$, where $V$ is the volume. The density of states in energy space can be obtained by changing variables from $p$ to $E$ using the relation $E = \frac{p^2}{2m}$, where $m$ is the mass of the electron. This gives:

$$
\rho(E) = \frac{V}{2\pi^2}\left(2m\right)^{\frac{3}{2}}\sqrt{E}.
$$

Finally, substituting the expressions for the matrix element and the density of states into Fermi's Golden Rule, we obtain:

$$
P_{fi} = \frac{2\pi}{\hbar}\left|\langle \psi_f | V | \psi_i \rangle\right|^2 \rho(E) = \frac{2\pi}{\hbar} \left|eE_0 \left(\frac{1}{\sqrt{\pi}}\right)\left(\frac{1}{a_0}\right)^{\frac{3}{2}} \int R_f(r)Y_{10}(\theta, \phi) (r \cos \theta) e^{-r/a_0} dV\right|^2 \frac{V}{2\pi^2}\left(2m\right)^{\frac{3}{2}}\sqrt{E}.
$$

$$
=\frac{V e^2E_0^2}{\hbar} \left|\left(\frac{1}{\sqrt{\pi}a_0}\right)^{\frac{3}{2}} \int R_f(r)Y_{10} r \cos \theta e^{-r/a_0} dV\right|^2 \left(2m\right)^{\frac{3}{2}}\sqrt{E}.
$$

This equation gives the transition rate from the ground state of the hydrogen atom to a continuum state as a function of the energy $E$ of the final state, the polarization of the electric field, and the angle of emission of the electron.
