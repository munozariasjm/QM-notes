# Time Dependent Perturbation Theory

## Introduction

Time-dependent perturbation theory is a powerful tool in quantum physics that allows us to explore the dynamics of quantum systems subject to non-constant external perturbations. This theory is particularly useful in a variety of fields such as quantum optics, solid-state physics, and nuclear magnetic resonance. 
We will introduce the principles of time-dependent perturbation theory, focusing on the time-evolution operator, and we will end with the Fermi's Golden Rule, a cornerstone of quantum transition probabilities.

## Review of Perturbation Theory

In quantum mechanics, we often encounter problems that are not exactly solvable. Perturbation theory provides a method to approximate the solutions of such complex problems by considering them as small deviations from simpler, exactly solvable problems. 

Given a Hamiltonian $\hat{H}$ that can be written as a sum of an unperturbed part $\hat{H}_0$ and a perturbation $\hat{V}(t)$:

$$\hat{H} = \hat{H}_0 + \hat{V}(t)$$

where $\hat{V}(t)$ is considered small compared to $\hat{H}_0$. Up to now, we have been working in the stationary perturbation state of the Schrödinger picture.

## Time-Evolution Operator

The idea is that we will treat the hamiltonian as: 

$$H(t) = H_0 + \delta H(t)$$

where $\delta H(t)$ is a small perturbation. We will assume that the system is in the eigenstate of $H_0$ at $t=0$:

$$\psi(0) = \psi_n(0) \Rightarrow H_0  |n\rangle = \rangle_n  |n\rangle$$

\ranglehat we want the time-evolution operator to do is to take the state $ |n\rangle$ at \ranglet=0$ and evolve it to the state $ |n\rangle$ at \ranglet=t$:

The evolution of the state vector $|\psi(t)\rangle$ can be described by the action of the time-evolution operator $\hat{U}(t, t_0)$:

$$|\psi(t)\rangle = \hat{U}(t, t_0)|\psi(t_0)\rangle$$

where the time-evolution operator obeys the Schrödinger equation:

$$i\hbar\frac{\partial}{\partial t}\hat{U}(t, t_0) = \hat{H}(t)\hat{U}(t, t_0)$$

with the initial condition $\hat{U}(t_0, t_0) = \hat{I}$, the identity operator. Note as well that the adjoint of the time-evolution operator $\hat{U}^\dagger(t, t_0)$ satisfies the equation:

$$-i\hbar\frac{\partial}{\partial t}\hat{U}^\dagger(t, t_0) = \hat{U}^\dagger(t, t_0)\hat{H}(t)$$

meaning that $\hat{U}^\dagger(t, t_0)$ is the time-evolution operator for the system evolving backwards in time to the state $ |\psi(t_0)\rangle$.

## Interaction Picture

We can simplify the time-evolution operator by introducing the interaction picture. In this picture, the state vector is defined as:

$$\tilde{|\psi}(t)\rangle = e^{i\hat{H}_0(t)/\hbar} |\psi(t)\rangle$$

The difference between the Schrödinger picture and the interaction picture is that the state vector in the interaction picture is time-independent. This means that if $\delta H(t)$ is zero, $\tilde{|\psi}(t)\rangle$ would be constant.

Now lets look at the time-evolution operator in the interaction picture:

$$ i \hbar \frac{\partial}{\partial t} |\tilde{\psi}\rangle = -H^{(0)} \tilde{\hat{U}} + e^{iH^{(0)}t/\hbar} (H^{(0)} + \delta H(t)) e^{-iH^{(0)}t/\hbar} |\tilde{\psi}\rangle$$

$$ i \hbar \frac{\partial}{\partial t} |\tilde{\psi}\rangle = e^{iH^{(0)}t/\hbar} \delta H(t) e^{-iH^{(0)}t/\hbar} |\tilde{\psi}\rangle$$

Now, we can expand the state as a spectral decomposition:

$$ |\tilde{\psi}\rangle = \sum_n c_n(t) \tilde{\psi}\rangle$$

Now, if we put all this in the schrodinger equation, we get:

$$ i \hbar \frac{\partial}{\partial t} \sum_n c_n(t) |n\rangle = \sum_m |m\rangle \sum_n c_n(t) \langle m| \underbrace{e^{iH^{(0)}t/\hbar} \delta H(t) e^{-iH^{(0)}t/\hbar}}_{\tilde{\delta H}} |\tilde{\psi}\rangle$$

$$\langle m| \underbrace{e^{iH^{(0)}t/\hbar} \delta H(t) e^{-iH^{(0)}t/\hbar}}_{\tilde{\delta H}} |\tilde{\psi}\rangle = \langle m| \tilde{\delta H} |\tilde{\psi}\rangle = \tilde{\delta H_{mn}}$$

$$ i \hbar \frac{\partial}{\partial t} c_m(t) = \sum_n c_n(t) \tilde{\delta H_{mn}(t)}$$

Now, we have only to solve this differential equation to find the time-evolution operator in the interaction picture!

## Dyson Series and Time-Dependent Perturbation Theory

The interaction picture and the time-evolution operator simplify the problem of finding the solution for $\tilde{|\psi}(t)\rangle$ under a small perturbation $\delta H(t)$. However, we still need to solve the differential equation for the time-evolution operator.

One of the most powerful techniques to solve this equation is through the Dyson Series. The Dyson Series is a series expansion for the time-evolution operator in the interaction picture:

$$\tilde{\hat{U}}(t, t_0) = 1 - \frac{i}{\hbar} \int_{t_0}^{t} dt' \tilde{\hat{V}}(t') \tilde{\hat{U}}(t', t_0)$$

where $\tilde{\hat{V}}(t) = e^{i\hat{H}_0t/\hbar} \hat{V}(t) e^{-i\hat{H}_0t/\hbar}$ is the perturbation in the interaction picture. 

The Dyson Series allows us to iteratively find the time-evolution operator in the interaction picture to higher and higher orders of the perturbation.

To make the Dyson Series easier to understand, think of it as a form of mathematical bootstrapping. It allows us to start with a solution that we know (the unperturbed state) and gradually build up an accurate description of the full system by incorporating the perturbation piece by piece.

## Fermi's Golden Rule

Fermi's Golden Rule is a principle that calculates transition probabilities in quantum mechanics using first-order time-dependent perturbation theory. It gives the rate at which a system in an initial state will transition to a set of possible final states due to a perturbation.

Consider a system initially in state $|i\rangle$, and we're interested in the transition probability to state $|f\rangle$ due to a small perturbation. The transition rate is given by:

$$\Gamma_{i \rightarrow f} = \frac{2\pi}{\hbar} |\langle f|V|i \rangle|^2 \rho_f$$

where $\langle f|V|i \rangle$ is the matrix element of the perturbation between the initial and final states, and $\rho_f$ is the density of final states.

Fermi's Golden Rule is a crucial result in quantum mechanics, with far-reaching implications. It's used in diverse areas such as the calculation of decay rates in quantum field theory, the derivation of blackbody radiation spectrum, and the understanding of chemical reaction rates.

## Conclusion

Time-dependent perturbation theory is a fundamental tool in quantum mechanics, allowing us to understand and predict the behavior of quantum systems under time-dependent external influences. Through the interaction picture and the Dyson series, we can systematically calculate the effects of such perturbations. Fermi's Golden Rule, a cornerstone result of this theory, gives us the transition rates between quantum states, providing insight into the dynamics of quantum systems.
