# Fundamental Concepts

### Kets, Bras, and Operators

The Dirac notation, which introduces the concepts of kets, bras, and operators, is a powerful language to express quantum mechanical ideas. In this chapter, we will delve into the details of these concepts, tracing the mathematical underpinnings and revealing key insights. 

#### Kets and Bras

We begin with a step into the abstract, setting aside the comfortable familiarity of three-dimensional space and venturing into the Hilbert space. In the realm of quantum mechanics, a physical state is represented by a vector in this infinite-dimensional complex vector space. These state vectors are what we refer to as 'kets'.

A ket is denoted as $|\psi\rangle$, where $\psi$ is a label that uniquely identifies a quantum state. 

Just as the shadows of three-dimensional objects provide different perspectives in two dimensions, we need a way to project our infinite-dimensional kets onto different axes to extract meaningful information. That's where the concept of 'bras' comes in. 

A bra is the Hermitian (conjugate) transpose of a ket and is denoted as $\langle\psi|$. This essentially means that if a ket $|\psi\rangle$ is represented as a column vector, its corresponding bra $\langle\psi|$ is a row vector, and the complex elements of the ket are replaced with their complex conjugates in the bra.

A key insight is that kets and bras aren't merely mathematical constructs, but are representative of physical reality. For instance, when a particle is in a state represented by $|\psi\rangle$, measurements conducted on the particle yield results related to the bra corresponding to the same state.

#### Inner and Outer Products

The inner and outer products are operations that you can perform with bras and kets. 

The inner product of two states $|\psi\rangle$ and $|\phi\rangle$ is denoted as $\langle\psi|\phi\rangle$. Mathematically, if $|\psi\rangle = (a_1, a_2, ..., a_n)$ and $|\phi\rangle = (b_1, b_2, ..., b_n)$, the inner product is given by:

$$\langle\psi|\phi\rangle = \sum_{i=1}^{n} a_i^* b_i$$

The inner product yields a complex number and provides a measure of the "overlap" between the two states. It's also important to note that the inner product of a state with itself ($\langle\psi|\psi\rangle$) gives the norm of the state.

The outer product is the reverse. The outer product of the states $|\psi\rangle$ and $|\phi\rangle$ is given by $|\psi\rangle\langle\phi|$. This results in a matrix that acts as a transformation operator.

#### Operators

Operators play a critical role in quantum mechanics as they are linked with observable quantities. Any measurement you can make on a quantum system corresponds to an operator. They act on kets to produce new kets and represent physical actions on the system.

Mathematically, operators are represented by matrices, with the caveat that in quantum mechanics, these matrices are often infinite-dimensional. Importantly, operators associated with observable quantities are Hermitian.

For instance, the momentum operator in one dimension is represented as $\hat{p} = -i\hbar\frac{d}{dx}$, where $i$ is the imaginary unit, $\hbar$ is the reduced Planck's constant, and $\frac{d}{dx}$ is the derivative with respect to position $x$.

When an operator acts on a ket, it provides a new state. For an operator $\hat{A}$ and a state $|\psi\rangle$, the action of the operator on the state is written as $\hat{A}|\psi\rangle$. 

When observables are measured, the outcomes are always real numbers, which correspond to the eigenvalues of the operator. If $|\psi\rangle$ is an eigenvector of $\hat{A}$, then we have

$$\hat{A}|\psi\rangle = a|\psi\rangle$$

where $a$ is a real number (the eigenvalue), indicating that if a measurement of the observable corresponding to $\hat{A}$ is made, the outcome will be $a$.

We now see how the concepts of kets, bras, and operators all interlink to form the foundations of quantum mechanics. Kets represent states, bras enable projection and calculation of inner products, and operators act on these states representing the measurements made on the system.

In the upcoming sections, we will explore these concepts further, delving into the specifics of particular operators and understanding how to work with these tools to extract meaningful information about quantum systems.

## Measurements, Observables, and Uncertainty Relations

We will now see how these concepts are utilized in the realm of measurements and observables, and how they give rise to the famous uncertainty relations of quantum mechanics.

#### Measurements and Observables

The act of measurement is central to quantum mechanics, and it's intrinsically tied to the concept of observables. In quantum physics, an observable corresponds to a physical property that can be measured, like energy, momentum, or position. Each observable is represented by a Hermitian operator, denoted as $\hat{A}$. 

The outcomes of the measurement of an observable are the eigenvalues of the corresponding operator. If $|\psi\rangle$ is a state of the quantum system and $\hat{A}$ is an observable with eigenvalues $a_i$, when a measurement is made, one of these eigenvalues will be observed. 

In mathematical terms, this can be stated as follows: suppose $|\psi\rangle$ can be expressed as a linear combination of the eigenvectors $|a_i\rangle$ of $\hat{A}$ such that 

$$|\psi\rangle = \sum_i c_i |a_i\rangle,$$

where $c_i$ are complex coefficients. Then, the probability $P(a_i)$ of obtaining the result $a_i$ upon measurement is given by $|c_i|^2$.

#### Post-Measurement State

A unique aspect of quantum measurement, which starkly contrasts with classical physics, is that the act of measurement changes the state of the system. Immediately after the measurement of an observable $\hat{A}$, the system will be in the eigenstate $|a_i\rangle$ corresponding to the eigenvalue $a_i$ that was measured. This is known as the collapse of the wavefunction or the reduction postulate. Mathematically, if the initial state was $|\psi\rangle = \sum_i c_i |a_i\rangle$, after obtaining the result $a_i$, the system's state will be $|a_i\rangle$.

#### Uncertainty Relations

The Heisenberg uncertainty principle is one of the most profound and fascinating aspects of quantum mechanics. It states that there are inherent uncertainties in the simultaneous measurement of two non-commuting observables. Most famously, this is seen in the position ($\hat{x}$) and momentum ($\hat{p}$) of a particle.

Mathematically, the uncertainty relation is expressed as:

$$\Delta x \Delta p \geq \frac{\hbar}{2},$$

where $\Delta x$ and $\Delta p$ are the uncertainties in the measurement of position and momentum, respectively, and $\hbar$ is the reduced Planck's constant.

This relation stems from the fact that the operators corresponding to these observables do not commute, i.e., $\hat{x}\hat{p} \neq \hat{p}\hat{x}$.

The principle fundamentally underscores the wave-particle duality of quantum entities and signifies a departure from the determinism of classical physics.

#### The Notion of Complementarity

At the heart of the uncertainty relations and the probabilistic nature of quantum mechanics is the principle of complementarity, introduced by Niels Bohr. It states that mutually exclusive classical concepts like particle and wave or position and momentum are complementary in the quantum world, meaning they together complete the description of quantum phenomena.

# Chapter 5: Change of Basis, Position, and Momentum

Change of basis and the concept of position and momentum are fundamental aspects of quantum mechanics. The former allows us to transform quantum states from one basis to another, making certain calculations easier. The latter two are examples of observable quantities in quantum mechanics, represented by operators in the corresponding Hilbert space.

#### Change of Basis

In quantum mechanics, different problems can be solved more easily in different bases. For example, the position representation is often useful when dealing with potentials that vary with position, while the momentum representation is convenient when dealing with free particles.

Mathematically, a change of basis corresponds to a unitary transformation. Given a ket $|\psi\rangle$ in the original basis, it can be transformed to a new basis using a unitary operator $\hat{U}$ as follows:

$$|\psi'\rangle = \hat{U} |\psi\rangle.$$

The bras transform with the conjugate transpose of the operator:

$$\langle\psi'| = \langle\psi| \hat{U}^\dagger.$$

Unitary operators preserve the inner product, meaning that the probabilities calculated in the new basis will be the same as in the old basis. This is crucial, as physical predictions must not depend on our choice of basis.

#### Position and Momentum

The position and momentum of a particle are two fundamental observables in quantum mechanics. In the position basis, the position operator $\hat{x}$ simply multiplies the state by the coordinate $x$, while in the momentum basis, the momentum operator $\hat{p}$ multiplies the state by the momentum $p$.

In the position basis, the momentum operator is represented as a derivative:

$$\hat{p} = -i\hbar \frac{d}{dx},$$

where $\hbar$ is the reduced Planck's constant, and $i$ is the imaginary unit.

The eigenstates of the position operator are delta functions at each point in space, and any wavefunction can be written as a superposition of these eigenstates. This corresponds to the particle being located at a specific point with 100% probability.

The momentum operator's eigenstates are plane waves with definite wavelength (and thus momentum). Any wavefunction can also be written as a superposition of these eigenstates, corresponding to the particle having a definite momentum.

As discussed in the previous chapter, position and momentum are complementary observables. Due to the Heisenberg uncertainty principle, they cannot both be known exactly at the same time. A state with a definite position has an entirely uncertain momentum, and vice versa.

#### The Fourier Transform and the Position-Momentum Duality

A deep connection exists between the position and momentum representations, mediated by the Fourier transform. The Fourier transform of a function gives you the coefficients when you expand the function in terms of plane waves (the momentum eigenstates), effectively changing from the position to the momentum basis. In quantum mechanics, this means you can go from a wavefunction in the position basis to one in the momentum basis using the Fourier transform.

This duality between position and momentum, and the role of the Fourier transform, is yet another manifestation of the wave-particle duality that lies at the heart of quantum mechanics.
