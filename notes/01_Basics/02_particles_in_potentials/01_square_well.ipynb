## Infinite Square Well

Lets think of a perticle inside a square well potential. The potential is given by:
\begin{equation}V(x)=\left\{\begin{array}{ll} 
-V_{0} & \text { for }
x \leq a \\ 
0 & \text { otherwise } 
\end{array}\right.\end{equation}

To solve this, we need to solve the time independent Schrodinger equation:

\begin{equation}\frac{d^{2} \psi}{d x^{2}}+\frac{2 m}{\hbar^{2}}\left[E-V(x)\right] \psi=0\end{equation}

If we consider the region where $V(x)=0$, the equation becomes:

$$\frac{d^{2} \psi}{d x^{2}}+\frac{2 m E}{\hbar^{2}} \psi=0$$

We will also introduce a new variable $k$:

$$k=\sqrt{\frac{2 m E}{\hbar^{2}}}$$

The solution to this equation must be of the form:

$$\psi(x)=A e^{i k x}+B e^{-i k x} \quad \rightarrow \quad \psi(x)=c_1 \sin (k x)+c_2 \cos (k x)$$

Now the matter is to use the boundary conditions to find the constants $c_1$ and $c_2$. The first boundary condition is that the wave function must vanish out at $x=0$ and $x=a$

$$\psi(0)=0 \quad \rightarrow \quad c_2=0$$

As well as the second boundary condition which is that the wave function must be continuous at $x=a$ and $x=b$:

$$\psi(a)=\psi(b) \quad \rightarrow \quad c_1 \sin (k a)=0$$

Following this logic, we arrive at a form for the wave function:

$$\psi_n = N \sin \left(\frac{n \pi x}{a}\right)$$

Where $N$ is a normalization constant and $n$ is an integer. To find the constant $N$, we use the normalization condition for the total probability:

$$\int_{-\infty}^{\infty}|\psi|^{2} d x=1$$

The energy is simply characterized by the $k$, which we found earlier:

$$E = \frac{\hbar^2 k^2}{2m} = \frac{\hbar^2 \pi^2 n^2}{2m a^2}$$

## Finite well

```julia
function finite_square_well(x, V0, a)
    if abs(x) < a
        return -V0
    else
        return 0
    end
end
x_axis = -2:0.01:2
y_axis = finite_square_well.(x_axis, 1, 1)
plot(x_axis, y_axis, label="Finite Square Well")
# Plot cartesian axes
plot!([minimum(x_axis), maximum(x_axis)], [0, 0], label="", color=:black)
plot!([0, 0], [minimum(y_axis), maximum(y_axis)], label="", color=:black)
```

To tackle this problem, we will use the energy with respect to the pottential:

$$\tilde E = V_0 - E \quad \Rightarrow \quad k^2 = \frac{2m}{\hbar^2} \tilde E$$

Note that we have to recover the infinite square well potential at $V_0 \rightarrow \infty$.
