# Chapter 1: Quantum States as Vectors in Hilbert Spaces

## 1.1 Introduction

In this chapter, we will introduce the fundamental concepts of quantum information by discussing quantum states, their representation as vectors in Hilbert spaces, and the mathematical framework that underlies quantum mechanics. By the end of this chapter, you should have a solid understanding of the basics of quantum information and be able to explore more advanced topics in the field.

## 1.2 Quantum States

Quantum states are the building blocks of quantum mechanics and quantum information. A quantum state represents the state of a quantum system and can be thought of as a generalization of classical states. In classical physics, a state is a point in phase space that uniquely defines the properties of a system. In contrast, a quantum state is described by a vector in a complex vector space called a Hilbert space.

### 1.2.1 Ket Notation

In quantum mechanics, we represent quantum states using Dirac's bra-ket notation. A quantum state is denoted by a ket, which is written as |ψ⟩, where ψ is a label for the state. The ket is a vector in the Hilbert space associated with the quantum system.

## 1.3 Hilbert Spaces

A Hilbert space is a complex vector space equipped with an inner product that allows the calculation of lengths and angles between vectors. The inner product also provides a notion of orthogonality, which is crucial for understanding quantum states and their superpositions.

### 1.3.1 Inner Product

The inner product of two vectors |ψ⟩ and |φ⟩ in a Hilbert space H is denoted as ⟨ψ|φ⟩. The inner product has the following properties:

1. Linearity: $$\langle\psi| (a |\phi_1\rangle + b |\phi_2\rangle) = a \langle\psi|\phi_1\rangle + b \langle\psi|\phi_2\rangle$$
2. Conjugate symmetry: $$\langle\psi|\phi\rangle = \langle\phi|\psi\rangle^*$$
3. Positivity: $\langle\psi|\psi\rangle \geq 0$, with equality if and only if |ψ⟩ = 0



### 1.3.2 Orthonormal Basis

An orthonormal basis is a set of vectors {|ψ_i⟩} in a Hilbert space H that satisfy the following conditions:

1. Normality: 
    $$\langle\psi_i|\psi_j\rangle = \delta_{ij}$$
2. Orthogonality:
    $$\langle\psi_i|\psi_j\rangle = 0, \text{ for } i \neq j$$

Given an orthonormal basis, any vector |ψ⟩ in H can be expressed as a linear combination of the basis vectors:

$$|\psi\rangle = \sum_i c_i |\psi_i\rangle$$

where the coefficients $c_i$ are complex numbers.

## 1.4 Qubits

The simplest quantum system is the qubit, which is the quantum analogue of a classical bit. A qubit has two basis states, usually denoted as |0⟩ and |1⟩, which form an orthonormal basis for the associated two-dimensional Hilbert space. A general qubit state can be written as:

$$|\psi\rangle = \alpha |0\rangle + \beta |1\rangle$$

where α and β are complex numbers such that:
$$\langle\psi|\psi\rangle = 1$$

Note that this quibit idea is applied for when the system is in a 2 dimensional Hilbert space. In general, the system can be in a Hilbert space of any dimension, for this, we define the **qudit**, as a generalization of the qubit:

$$|\psi\rangle = \sum_{i=0}^{d-1} c_i |i\rangle = \begin{bmatrix} c_0 \\ c_1 \\ \vdots \\ c_{d-1} \end{bmatrix}$$

where $|i\rangle$ is the ith basis state and $c_i$ are complex numbers such that:

$$\langle\psi|\psi\rangle = 1$$
